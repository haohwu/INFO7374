```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = 'https://www.timeout.com/boston/restaurants/the-34-best-boston-restaurants-right-now'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

session = requests.Session()
session.headers.update(headers)

try:
    # Send a GET request to the URL with session
    response = session.get(url, headers=headers, timeout=10)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the webpage
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Initialize the text content
        text_content = ""
        
        # Iterate through the elements in the body
        for element in soup.body.descendants:
            if element.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']:
                text_content += element.get_text() + '\n\n'
                
    else:
        text_content = f'Failed to retrieve the webpage. Status code: {response.status_code}'
except requests.exceptions.RequestException as e:
    text_content = f'Failed to retrieve the webpage. Error: {e}'

print(text_content)
```

```python
import os

# Directory to save the text file
directory = 'RAGdocs'

# Ensure the directory exists
os.makedirs(directory, exist_ok=True)

# Path to save the text file
txt_path = os.path.join(directory, "RestaurantsinBoston.txt")

# Write the text content to the text file
with open(txt_path, 'w', encoding='utf-8') as file:
    file.write(text_content)

print(f"Text file created successfully! Saved to {txt_path}")
```
